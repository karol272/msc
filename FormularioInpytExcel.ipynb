```python
# Registro de datos en excel desde una GUI en Tkinter 
# @autor: Diana Carolina / Alejandro Ospina
# Fecha: 07102023
# Youtube: https://www.youtube.com/c/MagnoEfren

from tkinter import Tk, Label, Button,Entry, Frame, END
import pandas as pd

ventana = Tk()
ventana.config(bg='black')
ventana.geometry('560x388')
ventana.resizable(0,0)
ventana.title('Guardar datos en Excel')
items1,descripcion_item1,bodega1,proveedor1,vendedor1 = [],[],[],[],[]

def agregar_datos():
	global items1, descripcion_item1, bodega1, proveedor1, vendedor1

	items1.append(ingresa_items.get())
	descripcion_item1.append(ingresa_descripcion_item.get())
	bodega1.append(ingresa_bodega.get())
	proveedor1.append(ingresa_proveedor.get())
	vendedor1.append(ingresa_vendedor.get())

	ingresa_items.delete(0,END)
	ingresa_descripcion_item.delete(0,END)
	ingresa_bodega.delete(0,END)
	ingresa_proveedor.delete(0,END)
	ingresa_vendedor.delete(0,END)

def guardar_datos():	
	global items1,descripcion_item1,bodega1,proveedor1,vendedor1
	datos = {'Items':items1,'Descripcion_item':descripcion_item1, 'Bodega':bodega1, 'Proveedor':proveedor1, 'Vendedor':vendedor1 } 
	nom_excel  = str(nombre_archivo.get() + ".xlsx")	
	df = pd.DataFrame(datos,columns =  ['Items', 'Descripcion_item', 'Bodega', 'Proveedor', 'Vendedor']) 
	df.to_excel(nom_excel)
	nombre_archivo.delete(0,END)

frame1 = Frame(ventana, bg='gray15')
frame1.grid(column=0, row=0, sticky='nsew')
frame2 = Frame(ventana, bg='gray16')
frame2.grid(column=1, row=0, sticky='nsew')

items = Label(frame1, text ='Item', width=10).grid(column=0, row=0, pady=20, padx= 10)
ingresa_items = Entry(frame1,  width=20, font = ('Arial',12))
ingresa_items.grid(column=1, row=0)

descripcion_item = Label(frame1, text ='Descripci√≥n del Item', width=10).grid(column=0, row=1, pady=20, padx= 10)
ingresa_descripcion_item = Entry(frame1, width=20, font = ('Arial',12))
ingresa_descripcion_item.grid(column=1, row=1)

Bodega = Label(frame1, text ='Bodega', width=10).grid(column=0, row=2, pady=20, padx= 10)
ingresa_bodega = Entry(frame1,  width=20, font = ('Arial',12))
ingresa_bodega.grid(column=1, row=2)

Proveedor = Label(frame1, text ='Proveedor', width=10).grid(column=0, row=3, pady=20, padx= 10)
ingresa_proveedor = Entry(frame1,  width=20, font = ('Arial',12))
ingresa_proveedor.grid(column=1, row=3)

Vendedor = Label(frame1, text ='Vendedor', width=10).grid(column=0, row=4, pady=20, padx= 10)
ingresa_vendedor = Entry(frame1, width=20, font = ('Arial',12))
ingresa_vendedor.grid(column=1, row=4)

agregar = Button(frame1, width=20, font = ('Arial',12, 'bold'), text='Agregar', bg='orange',bd=5, command =agregar_datos)
agregar.grid(columnspan=2, row=5, pady=20, padx= 10)

archivo = Label(frame2, text ='Ingrese Nombre del archivo', width=25, bg='gray16',font = ('Arial',12, 'bold'), fg='white')
archivo.grid(column=0, row=0, pady=10, padx= 10)

nombre_archivo = Entry(frame2, width=23, font = ('Arial',12),highlightbackground = "green", highlightthickness=4)
nombre_archivo.grid(column=0, row=1, pady=1, padx= 10)

guardar = Button(frame2, width=20, font = ('Arial',12, 'bold'), text='Guardar', bg='green2',bd=5, command =guardar_datos)
guardar.grid(column=0, row=2, pady=20, padx= 10)
ventana.mainloop()
```
